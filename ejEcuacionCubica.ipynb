Instrucción: Encontrar todas las raíces de la ecuación f(x) = 1/4(x^3 + 3x^2 - 6x - 8) usando el método de la bisección.
# Grafique. Modifique el número de iteraciones a 5. Pruebe los siguientes rangos de búsqueda: [-5, 3], [-4.7, 2.5], [-4.5, 2.7], [-3, -2].

```python
# Joseph Cruz, 12/05/2026
# Metodo de Biseccion con 5 iteraciones

import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 0.25 * (x**3 + 3*x**2 - 6*x - 8)

def biseccion(a, b, N0):
    if f(a) * f(b) >= 0:
        print(f"No hay cambio de signo en [{a}, {b}]")
        return None

    print(f"\nRango [{a}, {b}]")

    for i in range(1, N0 + 1):
        p = (a + b) / 2

        print(
            f"Iteracion {i}: "
            f"a={a:.6f}, "
            f"b={b:.6f}, "
            f"p={p:.6f}, "
            f"f(p)={f(p):.6f}"
        )

        if f(a) * f(p) > 0:
            a = p
        else:
            b = p

    return p

rangos = [
    [-5, -3],
    [-2, 0],
    [1, 3]
]

raices = []

print("Raices aproximadas:")

for r in rangos:
    raiz = biseccion(r[0], r[1], 5)

    if raiz is not None:
        raices.append(raiz)
        print(f"Raiz aproximada: {raiz:.6f}")

x = np.linspace(-6, 4, 1000)

plt.figure(figsize=(10, 5))
plt.plot(x, f(x), linewidth=2)
plt.axhline(0, linewidth=1)
plt.axvline(0, linewidth=1)

for raiz in raices:
    plt.plot(raiz, f(raiz), 'ro', markersize=8)

plt.grid(True)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Metodo de Biseccion")
plt.show()
```
